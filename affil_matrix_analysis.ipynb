```python
import pandas as pd
from tqdm import tqdm
```

```python
affil_mat = pd.read_csv('affiliation_matrix_combined_final.csv',index_col = 0)
```

```python
candidates = affil_mat.columns
candidates
```

```python
def average_affiliation(df, candidate):
    df_candidate = df[df[candidate] == 1]
    avg_affiliation = (df_candidate.sum(axis= 1)-1).mean()
    return avg_affiliation
```

```python
candidate_avg_affil = []
for candidate in candidates:
    print(f'{candidate}: ',average_affiliation(affil_mat, candidate))
    candidate_avg_affil.append(average_affiliation(affil_mat, candidate))

candidate_avg_affil_df = pd.DataFrame(candidate_avg_affil, index=candidates, 
                                      columns=["Average Number of Affiliations"])
candidate_avg_affil_df.to_csv("candidate_avg_affil.csv")
```

```python
affil = pd.DataFrame(0, index=candidates, columns=candidates)
for candidate in tqdm(candidates):
    df_candidate = affil_mat[affil_mat[candidate] == 1]
    for user in list(df_candidate.index):
        if df_candidate.loc[user].sum() == 0:
            continue
        for candidate_2 in list(df_candidate.columns):
            if df_candidate.loc[user][candidate_2] == 1:
                affil.loc[candidate][candidate_2]+= 1
#             affil.loc[candidate][candidate_2]+=1
```

```python
affil.to_csv('candidate_follower_overlap_matrix.csv')
```

```python
# affil_norm = affil.copy()
affil_norm = pd.DataFrame(0.0, index=candidates, columns=candidates)
for candidate in list(affil.index):
    candidate_total = affil.loc[candidate][candidate]
#     print(candidate_total)
    for candidate_2 in list(affil.index):
        ratio = round(affil.loc[candidate][candidate_2]/candidate_total,3)
#         affil_norm.set_value(candidate,candidate_2, ratio)
        affil_norm.at[candidate,candidate_2] = ratio
```

```python
affil_norm.to_csv('candidate_follower_overlap_matrix_norm.csv')
```
